## Q5

According to Kepler, celestial objects orbit around one of their foci in the shape of an ellipse.  Thus, given the information from the Halley Commet wiki, we can find an equation for its orbit around the sun.

Semi-major axis (a): 17.834 AU

apahelion = 0.586 AU

Finding c

    C = a - apahelion

```python
c = 17.834 - 0.586
c
```

Finding b

$$ b^2 + c^2 = a^2 $$ 


```python
a = 17.834

b = ((a ** 2) - (c ** 2)) ** 0.5
b
```

Since we have a and b, we can write the equation of the ellipse

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$

or

$$y = \pm b\sqrt{(1 - \frac{x^2}{a^2})}$$

to make sure that we have a foci (the sun) at the origin, we shift by c

$$y = \pm b\sqrt{(1 - \frac{(x-c)^2}{a^2})}$$

```python
def h_path(a,b,c,x):
    return b * ((1 - ((x - c)**2 / a**2)) ** 0.5)
```

```python
x = np.linspace((c - a), (c + a), num=100) #start at the apahelion and go to the perihelion
plt.plot(x, h_path(a,b,c,x), color = 'blue')
plt.plot(x, -h_path(a,b,c,x), color = 'blue')
plt.plot(0,0,'oy')
plt.title("Halley's Commet path around the sun")
plt.xlabel('Displacement from the Sun (AU)')
plt.ylabel('Displacement from the Sun (AU)')
```
